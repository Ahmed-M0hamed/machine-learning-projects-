```python
import tensorflow as tf 
import numpy as np 
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# fetch the transcript by the video id 
video_path = "fGx6K90TmCI" 
y_t = YouTubeTranscriptApi()
scripts = y_t.fetch(video_path) 
for snippet in scripts:
    print(snippet.text)
```

# using first method for song embedding 
- we use the snipets as they are 
- we toknize and pad them 
- we fit them to the model 

```python
train_texts = []
for snippet in scripts: 
    train_texts.append(snippet.text.split()[1:-1]) 
```

```python
train_text  = np.array(train_texts)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer() 
tokenizer.fit_on_texts(train_text)
```

```python
tokenized_text = tokenizer.texts_to_sequences(train_text) 
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_tokens = pad_sequences(tokenized_text) 
```

```python
train_data = []
train_labels = []
for token in padded_tokens : 
    train_token = token[:-1] 
    train_label = token[-1]
    train_data.append(train_token) 
    train_labels.append(train_label) 
```

```python
train_data= np.array(train_data) 
train_labels = np.array(train_labels) 
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
model = tf.keras.Sequential([
    tf.keras.Input(shape = (12 ,)) , 
    tf.keras.layers.Embedding(198 , 64) , 
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20)) , 
    tf.keras.layers.Dense(56 , activation = 'softmax') 
])
model.compile(loss= 'categorical_crossentropy' , optimizer ='adam' , metrics= ['accuracy']) 
model.fit(train_data,  train_labels , epochs=500 ) 
```

```python
e = model.evaluate(train_data , train_labels)
```

```python
test_text= ['Arf arf']
t_text = tokenizer.texts_to_sequences(test_text) 
test_padded = pad_sequences(t_text , maxlen=12 ) 
```

```python
seed_text = 'Arf arf'
for i in range(100) : 
    t_text =  tokenizer.texts_to_sequences([seed_text])[0] 
    test_padded = pad_sequences([t_text] , maxlen=12 )  
    prediction = model.predict(test_padded) 
    p = np.argmax(prediction, axis=-1)[0] 
    if p != 0:

        output_word = tokenizer.index_word[p]
        seed_text += " " + output_word
print(seed_text)
```

# second method 
- we take the song as one text 
- we make tokens of each n words togather 
- we toknize them and pad them 

```python
whole_song = ''
for s in scripts.snippets : 
    words = s.text.split()[1:-1] 
    whole_song += ' '.join(words)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer() 
tokenizer.fit_on_texts(whole_song.split())
```

```python
whole_song
```

```python
input_sequences = []
for i in range(1, len(whole_song.split())) : 
        sequence = ' '.join(whole_song.split()[:i+1])
        input_sequences.append(sequence)
```

```python
input_sequences[0]
```

```python
tokenized_seqences = tokenizer.texts_to_sequences(input_sequences)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_seqences = pad_sequences(tokenized_seqences) 
```

```python
padded_seqences[0]
```

```python
train_data = padded_seqences[: , :-1] 
train_labels = padded_seqences[: , -1]
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape = (582 ,)) , 
    tf.keras.layers.Embedding(214 , 64) , 
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(50 , return_sequences=True)) , 
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20)) , 
    tf.keras.layers.Dense(215 , activation = 'softmax') 
])
model.compile(loss= 'categorical_crossentropy' , optimizer ='adam' , metrics= ['accuracy']) 
model.fit(train_data,  train_labels , epochs=500 ) 
```

```python
seed_text = 'ArfYeahUhYeahYeahUhYeah'
for i in range(30) : 
    t_text =  tokenizer.texts_to_sequences([seed_text])[0] 
    test_padded = pad_sequences([t_text] , maxlen=582 )  
    prediction = model.predict(test_padded) 
    p = np.argmax(prediction, axis=-1)[0] 
    if p != 0:

        output_word = tokenizer.index_word[p]
        seed_text += " " + output_word
print(seed_text)
```
